```python
game = chess.pgn.read_game(io.StringIO(tal))
record = pipeline.parse_game(game, 201089773, engine, chess.engine.Limit(depth=14), chess.engine.INFO_SCORE)
```

```python
# Plot area chart of WDL over moves
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(record))
y_win = np.zeros(len(record))
y_draw = np.zeros(len(record))
y_loss = np.zeros(len(record))

for i, rec in enumerate(record):
    y_win[i] = rec["winning_chance"]
    y_draw[i] = rec["drawing_chance"]
    y_loss[i] = rec["losing_chance"]

plt.stackplot(x, y_win, y_draw, y_loss, labels=['Win', 'Draw', 'Loss'], colors=['#4CAF50', '#FFC107', '#F44336'])
plt.xlabel('Move Number')
plt.ylabel('Probability')
plt.title('WDL Probabilities Over Moves')
plt.legend(loc='upper right')
plt.show()
```
